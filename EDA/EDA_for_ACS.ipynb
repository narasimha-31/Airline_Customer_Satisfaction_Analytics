```python
#above we are meesed up and i mean we are not able to reuse the labels so now we will make the code smaller by reusing the labels by defining them at once and use forever 
```

```python
def apply_labels(title,x_label,y_label):
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
```

```python
sns.countplot(data=df,x='class',color='green')
apply_labels("Passenger Seating Class","Class","No of Passengers")
plt.show()
```

```python
sns.countplot(data=df,x='age group',hue='Overall',palette='deep')
```

```python
sns.countplot(data=df,x='age group',hue='Overall',palette='deep')
plt.title("Satisfaction Distribution Across Age Groups")
plt.xlabel("Age Group")
plt.ylabel("Passenger Count")
plt.legend(title="Satisfaction Level")
plt.show()
```

```python
ax=sns.countplot(data=df,x='age group',hue='Overall',palette='deep')
plt.title("Satisfaction Distribution Across Age Groups")


plt.xlabel("Age Group")
plt.ylabel("Passenger Count")
plt.legend(title="Satisfaction Level", loc='upper left', bbox_to_anchor=(1, 1))
for container in ax.containers:
    ax.bar_label(container, padding=3, fontsize=9)

plt.show()
```
